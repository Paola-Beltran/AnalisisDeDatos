```python
# filtrar datos por dos condiciones
# No es tan eficiente para volumenes de datos grandes
df_sales[(df_sales.STATE == 'NY') & (df_sales.QTR_ID == 1)]
```

```python
# se utiliza .query es mas eficiente para filtrar grandes volumnes de datos
df_sales.query('STATE == "NY" and QTR_ID == 1')
```

```python
# agrupar para saber cuantos registros hay dentro de cada pais
df_sales[['STATE', 'COUNTRY']].groupby('COUNTRY').count()
```

```python
# para dejar nuevamente Country como una columna y no como indice se usa reset_index, para asignar el indice

df_grp = df_sales[['STATE', 'COUNTRY']].groupby('COUNTRY').count()
print(type(df_grp))
df_grp.reset_index()
```
